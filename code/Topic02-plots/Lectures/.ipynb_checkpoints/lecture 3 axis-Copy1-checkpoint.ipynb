# subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

we can create subplots by passing parameters into the subplot function.
it will return a fugure and a 2d array of axes

parameters 
`ncols`, `nrows`

you can set the size using `figsize`

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(5.5, 4.5))


for row in range(2):
    for col in range(2):
        axs[row, col].annotate(f'axs[{row}, {col}]', (0.5, 0.5),
                               ha='center',  fontsize=18,
                               color='darkgrey')
fig.suptitle('plt.subplots()')

plt.show()
```

### Example

```python
data = [
    [2,4,5,6,3,5,6,7],
    [9,8,7,6,5,4,3,2],
    [1,2,3,4,5,6,7,8],
    [5,5,5,5,5,5,5,5]
]
x = 0.5 + np.arange(8)
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(5.5, 3.5))
# add an artist, in this case a nice label in the middle...
count = 0
for row in range(2):
    for col in range(2):
        axs[row, col].plot(x,data[count])
        count += 1
fig.suptitle('plt.subplots()')

plt.show()
```
